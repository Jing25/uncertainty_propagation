```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm
```

```python
# Plot between -10 and 10 with .001 steps.
x_axis = np.arange(-5, 5, 0.01)
# Mean = 0, SD = 2.
plt.plot(x_axis, norm.pdf(x_axis,0,2))
```
